```python

# Extract precision and recall
precision = results.results_dict['metrics/precision(B)']
recall = results.results_dict['metrics/recall(B)']

# Calculate the accuracy-like metric (you can use precision or an average of precision and recall)
accuracy_like_metric = precision

# Alternatively, you can use the F1 Score as an accuracy-like metric
if (precision + recall) > 0:
    f1_score = 2 * (precision * recall) / (precision + recall)
else:
    f1_score = 0.0

# Print the results
print(f"Precision (as Accuracy-like metric): {accuracy_like_metric:.4f}")
print(f"F1 Score (as Accuracy-like metric): {f1_score:.4f}")
```
